## Importing packages

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

## Reading and showing the  image

```python
img = cv2.imread('img.webp')
plt.imshow(img)
```

## Applying some functions like gray scale , threshold, kernal and dilation the image

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_OTSU|cv2.THRESH_BINARY_INV)
kernal = cv2.getStructuringElement(cv2.MORPH_RECT,(20,20))
dilation = cv2.dilate(thresh,kernal,iterations=1)
```

## Finding contours 

```python
contours, hierarchy = cv2.findContours(dilation,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
img1 = img.copy()
```

## By using those contours we can crop the text and gave to pytesseract

```python
with open('extracted.txt','w+') as f:
    for cnt in contours:
        x,y,w,h = cv2.boundingRect(cnt)
        rect = cv2.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),2)
        cropped = img1[y:y+h,x:x+w]
        text = pytesseract.image_to_string(cropped)
        print(text)
        f.write(text)
```
